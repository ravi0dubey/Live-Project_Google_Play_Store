### Handling the Size Features. It has K and M in it.

```python
# Checking all the unique values in the Size feature

df['Size'].unique()
```

##### Remove all characters from size and convert it to float

```python
df['Size'] = df['Size'].str.replace('k', " ")
df['Size'] = df['Size'].str.replace('M', "000")
df['Size'] =df['Size'].replace("Varies with device",np.nan)
df['Size'] = df['Size'].astype(float)
```

```python
df.head(5)
```

##### Convert Megabytes which were in decimal, even after adding '000' in the end, decimal values will get truncated once we change the datatype from string to float . So we need to convert the Megabytes into Kilo bytes then convert all to mega bytes. So that all size lies in Sync

```python
for i in df['Size']:
    if i < 10:
        df['Size']= df['Size'].replace(i, i*1000)      
df['Size'] = df['Size']/1000
```

```python
df.head(5)
```

### Handling Installs and Price Feature. 
    It has "+" in it which needs to be changed to spaces and "," needs to be removed before converting entire feature into     Int types

```python
### We can use below if we have just one or 2 columns to change with lesser values

# df['Installs'] = df['Installs'].str.replace("+","" )
# df['Installs'] = df['Installs'].str.replace(",","" )
# df['Installs'] = df['Installs'].astype(int)

# df['Price'] = df['Price'].str.replace("$"," " )
# df['Price'] = df['Price'].astype(float)
```

```python
### We Should be using it if we have more columns and more unique values to replace

chars_to_remove = [',',"$","+"]
cols_to_clean = ['Installs','Price']

for i in chars_to_remove:
    for j in cols_to_clean:
        df[j]= df[j].str.replace(i,"")

df['Installs'] = df['Installs'].astype(int)
df['Price'] = df['Price'].astype(float)
```

### Convert the 'Last Updated' column to Last Updated format

```python
# df['Last Updated']= df['Last Updated'].astype('datetime64[ns]')
##                    OR

df['Last Updated']= pd.to_datetime(df['Last Updated'])
```

```python
# Storing Individual day, month and year in separate Columns

df['Day']=df['Last Updated'].dt.day
df['Month']=df['Last Updated'].dt.month
df['Year']=df['Last Updated'].dt.year
```
