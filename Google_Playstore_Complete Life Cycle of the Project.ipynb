### Handling the Size Features. It has K and M in it.

```python
# Checking all the unique values in the Size feature

df['Size'].unique()
```

##### Remove all characters from size and convert it to float

```python
df['Size'] = df['Size'].str.replace('k', " ")
df['Size'] = df['Size'].str.replace('M', "000")
df['Size'] =df['Size'].replace("Varies with device",np.nan)
df['Size'] = df['Size'].astype(float)
```

```python
df.head(5)
```

##### Convert Megabytes which were in decimal, even after adding '000' in the end, decimal values will get truncated once we change the datatype from string to float . So we need to convert the Megabytes into Kilo bytes then convert all to mega bytes. So that all size lies in Sync

```python
for i in df['Size']:
    if i < 10:
        df['Size']= df['Size'].replace(i, i*1000)      
df['Size'] = df['Size']/1000
```

```python
df.head(5)
```

### 4.c Random Sample Imputation

```python
# replace df with df_copy1 to have cleaned data
df= df_copy1
df.head()
```

```python
df[df['Size'].isnull()].index
```

```python
df['Size'].dropna().sample(1695)
```

```python
df['Size'].dropna().sample(df['Size'].isnull().sum())
```

```python
def Random_Sample_imputation(feature):
    random_sample=df[feature].dropna().sample(df[feature].isnull().sum())               
    random_sample.index=df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature]=random_sample
```

```python
for col in df:
    Random_Sample_imputation(col)
```

```python
df[df['Size'].isnull()]
```

```python
print('Original Size Variance', df_copy1['Size'].var())
print('Size Variance After Random imputation', df['Size'].var())


print('Original Rating Variance', df_copy1['Rating'].var())
print('Rating Variance After Random imputation', df['Rating'].var())
```

```python
fig= plt.figure()

# density plot using seaborn library
fig, axs = plt.subplots(2, 2, figsize=(15, 7))

df_copy1['Size'].plot.density(color='red',ax=axs[0, 0],alpha=0.5,label='Size')
df['Size'].plot.density(color='green',ax=axs[0, 0],alpha=0.5,label='Size')
df_copy1['Rating'].plot.density(color='red',ax=axs[0, 1],alpha=0.5,label='Rating')
df['Rating'].plot.density(color='green',ax=axs[0, 1],alpha=0.5,label='Rating')

df_copy1['Size'].hist(bins=50,ax=axs[1, 0],density=True,figsize=(12,12),color='red')
df['Size'].hist(bins=50,ax=axs[1, 0],density=True,figsize=(12,12),color='green', alpha=0.8)
df_copy1['Rating'].hist(bins=50,ax=axs[1, 1],density=True,figsize=(12,12),color='red')
df['Rating'].hist(bins=50,ax=axs[1, 1],density=True,figsize=(12,12),color='green', alpha=0.8)

plt.show()
```
