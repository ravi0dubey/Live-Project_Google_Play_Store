### 2 . Z-score method
#### Z-score:
    The number of standard deviations away from the mean that a particular observation is.
    A negative Z-score means an observation is below the mean.
    while a positive Z-score means means it above the mean.
    The further away from 0 the Z-Score is, the further away from the mean your observation is.

```python
# Function to detect outliers Using Quartiles with Q1 at 10% and Q3 at 90%

# It will return lower and upper fence

def outlier_thresholds(dataframe, variable):
    quartile1 = dataframe[variable].quantile(0.10)
    quartile3 = dataframe[variable].quantile(0.90)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    print(f"up_limit : {up_limit}")
    print(f"low_limit : {low_limit}")
    return low_limit, up_limit
```

```python
## function to replace outliers with lower limit and Upper limit

def replace_with_thresholds(dataframe, numeric_columns):
    for variable in numeric_columns:
        low_limit, up_limit = outlier_thresholds(dataframe, variable)
        print(f"dataframe.loc[(dataframe[variable] < low_limit), variable]: {dataframe.loc[(dataframe[variable] < low_limit), variable]}")
        dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
        print(f"dataframe.loc[(dataframe[variable] > up_limit), variable]: {dataframe.loc[(dataframe[variable] > up_limit), variable]}")
        dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

### 3. Interquartile Range Method( IQR ) method


```python
for col in num_df.columns:
    plt.figure(figsize=(12,4))
    sns.boxplot(df[col])
    plt.title(col)
```

```python
# Take backup of the dataframe
df1 = df.copy()


# Create a function for removal of Outliers. Replace all values lower than lower_limit, with the lower_limit 
# and all values greater than upper limit, replace it with upper limit values


def remove_outliers_IQR(col):
    # Finding the IQR
    percentile25 = df1[col].quantile(0.25)
    percentile75 = df1[col].quantile(0.75)
    print("percentile25",percentile25)
    print("percentile75",percentile75)
    iqr = percentile75 - percentile25
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    print("Upper limit",upper_limit)
    print("Lower limit",lower_limit)
    df1[col] = np.where(df1[col]>upper_limit, upper_limit, np.where(df1[col]<lower_limit,lower_limit,df1[col]))
#     return df1[df1[col] > upper_limit]
```

```python
remove_outliers_IQR('Size')
```

```python
def create_comparison_plot(df,df1,column):
    # Comparing
    plt.figure(figsize=(16,8))
    plt.subplot(2,2,1)
    sns.distplot(df[column])

    plt.subplot(2,2,2)
    sns.boxplot(df[column])

    plt.subplot(2,2,3)
    sns.distplot(df1[column])

    plt.subplot(2,2,4)
    sns.boxplot(df1[column])

    plt.show()
```
